# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
```

# line plot

```python
#ðŸ”¹ Line Color & Style
plt.plot(x,y,color='green',linestyle='--',linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('simple-line-plot')
plt.show()
```

```python
# ðŸ”¹ Marker Style
plt.plot(x,y,marker='o', markersize=8,markerfacecolor='blue')
```

```python
# ðŸ”¹ Axis Limits
plt.plot(x,y)
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

```python
# ðŸ”¹ Add Legend
plt.plot(x, y, label='Growth')
plt.legend()
```

```python
plt.plot(x, y, label='Growth')
plt.xlim(0,6)
plt.legend()
plt.text(3,4,'midpoint',fontsize=12)
```

# Bar plot

```python
plt.bar(x,y ,linestyle = '-', width=0.5, color = 'red' ,label='Growth')
plt.title('Bar Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(ticks=x, labels=["A", "B", "C", "D", "E"], rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show()
```

# scatter plot

```python
c = np.array([10, 20, 30, 40, 50]) 
plt.scatter(x,y, c=c, cmap='viridis', marker='o', s = 20, label='Data Points')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(ticks=x)
plt.yticks(ticks=y)
plt.text(x[np.argmax(y)],y.max(), "Max", fontsize=12, color='red')
plt.text(x[np.argmin(y)],y.min(), "Min", fontsize=12, color='blue')
plt.grid()
plt.legend()
plt.colorbar(label="Intensity")
plt.show()
```

# Histogram

```python
data1 = np.random.randn(100)
data2 = np.random.randn(100)  # Generate random data
plt.hist(data1,bins = 10, color='purple', edgecolor='black', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
error = np.array([0.5, 1, 0.2, 0.3, 0.7])
plt.errorbar(x, y, yerr=error, fmt='o--', color='red')
```

# Area plot

```python
plt.style.use('Solarize_Light2')
plt.fill_between(x, y, color='skyblue', alpha=0.5, label='Sine Area')
plt.plot(x, y, color='blue')  # Line on top of the filled area
plt.title("Area Plot (fill_between)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.style.use('seaborn-v0_8-white')
plt.grid(True)
plt.legend()
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='orange')
plt.title("Area Between Two Curves")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.style.use('seaborn-v0_8-white')
plt.grid(True)
plt.show()
```
